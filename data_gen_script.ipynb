```python
# Function to generate random datetime within specified range
def generate_random_datetime():
    # Define date range from Jan 2020 to Dec 2022
    start_date = datetime(2020, 1, 1)
    end_date = datetime(2022, 12, 31)
    
    # Generate random date within the range
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    
    # Generate random time within store opening hours (9am to 10pm)
    random_time = random.randint(9, 21)  # Hours from 9am to 9pm
    random_minute = random.randint(0, 59)
    
    # Combine date and time
    random_datetime = random_date.replace(hour=random_time, minute=random_minute)
    
    return random_datetime

# Apply the function to create a new column for datetime
transaction_df['datetime'] = transaction_df.apply(lambda _: generate_random_datetime(), axis=1)

# Extract date and time from the datetime column
transaction_df['date'] = transaction_df['datetime'].dt.strftime('%m/%d/%Y')
transaction_df['time'] = transaction_df['datetime'].dt.strftime('%H:%M')

# Drop the datetime column if not needed anymore
transaction_df.drop(columns=['datetime'], inplace=True)
```

### Customer table

```python
customer_df = pd.read_csv(r"C:\Users\omari\Documents\GitHub\Megastore-Expansion-Initiative\unclean_data\customer_dim_n.csv")

customer_df.head()
```

```python
# Ensure the customer_df has the customer_id column
if 'customer_id' not in customer_df.columns:
    raise ValueError("CSV file must contain a 'customer_id' column")

# Number of customers
num_customers = customer_df.shape[0]

# Gender distribution
genders = ['f', 'm']
gender_probs = [0.6, 0.4]

# Age groups distribution
age_groups = [
    (19, 30),
    (31, 50),
    (51, 65),
    (66, 70)
]
age_group_probs = [0.25, 0.45, 0.25, 0.05]

# Employment status distribution per age group
employment_statuses = {
    (19, 30): ['student', 'employed', 'business owner', 'unemployed'],
    (31, 50): ['employed', 'business owner', 'unemployed', 'student'],
    (51, 65): ['employed', 'business owner', 'unemployed', 'retired'],
    (66, 70): ['retired', 'business owner']
}
employment_probs = {
    (19, 30): [0.6, 0.2, 0.03, 0.17],
    (31, 50): [0.7, 0.05, 0.1, 0.15],
    (51, 65): [0.5, 0.3, 0.1, 0.1],
    (66, 70): [0.85, 0.15]
}

# Marital status distribution per age group
marital_statuses = {
    (19, 30): ['married', 'single', 'divorced'],
    (31, 50): ['married', 'single', 'divorced'],
    (51, 65): ['married', 'single', 'divorced'],
    (66, 70): ['married', 'single', 'widowed']
}
marital_probs = {
    (19, 30): [0.3, 0.65, 0.05],
    (31, 50): [0.6, 0.25, 0.15],
    (51, 65): [0.7, 0.2, 0.1],
    (66, 70): [0.4, 0.25, 0.35]
}

# Verify that all probability distributions sum to 1
def verify_probabilities(probs):
    return np.isclose(sum(probs), 1.0)

assert verify_probabilities(gender_probs), "Gender probabilities do not sum to 1"
assert verify_probabilities(age_group_probs), "Age group probabilities do not sum to 1"
for age_group in age_groups:
    assert verify_probabilities(employment_probs[age_group]), f"Employment probabilities for age group {age_group} do not sum to 1"
    assert verify_probabilities(marital_probs[age_group]), f"Marital probabilities for age group {age_group} do not sum to 1"

# Assign gender
customer_df['sex'] = np.random.choice(genders, size=num_customers, p=gender_probs)

# Assign age and related demographics
ages = []
employment_status = []
marital_status = []

for _ in range(num_customers):
    age_group_idx = np.random.choice(len(age_groups), p=age_group_probs)
    age_group = age_groups[age_group_idx]
    age = np.random.randint(age_group[0], age_group[1] + 1)
    ages.append(age)
    
    emp_status = np.random.choice(employment_statuses[age_group], p=employment_probs[age_group])
    employment_status.append(emp_status)
    
    marital_stat = np.random.choice(marital_statuses[age_group], p=marital_probs[age_group])
    marital_status.append(marital_stat)

customer_df['age'] = ages
customer_df['employment_status'] = employment_status
customer_df['marital_status'] = marital_status
```
